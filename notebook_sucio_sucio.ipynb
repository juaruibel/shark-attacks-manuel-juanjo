Comprobamos los valores unicos en la columna fatal

```python
#Funcion para definir que solo queremos valores 'Y' o 'N'

def clean_fatal(valor):
    if valor == 'Y':
        return 'Y'
    elif valor == 'N':
        return 'N'
    else:
        return 'UNKNONW'
```

```python
df_shark_attacks['fatal'] = df_shark_attacks['fatal'].apply(clean_fatal)
df_shark_attacks['fatal'].value_counts()
```

Comprobamos como estan escritos las principales nombres de la especies

```python
df_shark_attacks['species'] = df_shark_attacks['species'].fillna('Unknown').astype(str)
```

Rellenamos los nulos con Unknown y a tipo str

```python
#Funcion para mejorar los valores dentro de la tabla species

def categoria_species(valor):
    valor = valor.lower()  #convertimos todo a minuscula

    if 'white' in valor:
        return 'White Shark'
    elif 'tiger' in valor:
        return 'Tiger Shark'
    elif 'bull' in valor:
        return 'Bull Shark'
    elif 'hammerhead' in valor:
        return 'Hammerhead Shark'
    elif 'shark' in valor:
        return 'Other Shark'
    else:
        
        return 'Unknown'
    
df_shark_attacks['species'] = df_shark_attacks['species'].apply(categoria_species)
```
